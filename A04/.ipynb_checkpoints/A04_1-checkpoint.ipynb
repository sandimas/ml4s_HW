```python
mydigits = []
yval = []
MNIST_len = 28

def gauss_2d(x,y,sigma):
    return np.exp(-(x**2+y**2)/(2*sigma**2))

def downsample(_image,_new_len):
    _sigma = 0.01
    _outimage = np.zeros((_new_len,_new_len))
    n_pix=np.shape(_image)[0]
    binratio = n_pix/_new_len
    relative_coord = []
#     for j in range(n_pix):
#         relative_coord.append(j*_new_len/n_pix)
    image = np.pad(_image,(2,2),constant_values=0)
    for i in range(5,6):
        for j in range(5,6):
            for i_old in range(n_pix+4):
                for j_old in range(n_pix+4):
                    relx = i/_new_len - (i_old-2)/n_pix
                    rely = j/_new_len - (j_old-2)/n_pix
                    g2 = gauss_2d(relx,rely,_sigma)
                    print(g2)
#                     _outimage[i,j] += g2
    print(_outimage[5,5])
    return _outimage
    

for i in range(1):
    yval.append(i)
    #read
    image = plt.imread(f'./my_digits/{i}.png')[:,:,1]
    image -= np.min(image)
    image /= np.max(image)
    image = 1-image
    plt.matshow(image,cmap='binary')
    print(f'reading {i}')
    #rectangularize
    shape = np.shape(image)
    xlen = shape[0]
    ylen = shape[1]
    dif = xlen-ylen
    if (dif > 0): #x > y
        padl = dif//2
        padr = dif - padl
        image = image[padl:xlen-padr,:]
    elif (dif < 0):
        padl = -dif//2
        padr = -dif - padl
        image = image[:,padl:ylen-padr]
    
    # downsample
    image = downsample(image,MNIST_len)
   
    
    mydigits.append(image)
print("done")
```
